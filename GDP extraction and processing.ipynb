```python
import numpy as np
import pandas as pd

# You can also use this section to suppress warnings generated by your code:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')
```

```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
# Extract tables from webpage using Pandas. Retain table number 3 as the required dataframe.

x = pd.read_html(URL)

df = x[3]
df.head() ##  It is already a dataframme

# Replace the column headers with column numbers
df.columns = range(df.shape[1])

#print(df.head())

# Retain columns with index 0 and 2 (name of country and value of GDP quoted by IMF)

new_df = df[[0, 2]]
new_df.head()


# Retain the Rows with index 1 to 10, indicating the top 10 economies of the world.
new_df = new_df.iloc[1:11, :]
new_df.head()
# Assign column names as "Country" and "GDP (Million USD)"

new_df.columns = ['Country', 'GDP (Million USD)']

new_df.head()
```

```python
## Converting it to billion
```

```python
new_df.info()
```

```python
## Change data type of GDP column

new_df['GDP (Million USD)'] = new_df['GDP (Million USD)'].astype(int)
```

```python
new_df.info()
```

```python
## Converting to Billions divide by 1000
new_df['GDP (Million USD)'] = new_df['GDP (Million USD)']/1000
```

```python
new_df[['GDP (Million USD)']]
```

```python
## Rounding upto 2 decimals
new_df[['GDP (Million USD)']] = np.round(new_df[['GDP (Million USD)']], 2)
new_df[['GDP (Million USD)']]
```

```python
## rename columns
## When using df.rename should give  the names as a dictionary
new_df.rename(columns = {'GDP (Million USD)': 'GDP (Billion USD)'})
```

```python
## Copying to csv

new_df.to_csv('./filename.csv', index = False)
```
